```python
import pandas as pd
#Importando a biblioteca pandas do Python para a leitura do arquivo csv.
#fonte World Health Organization
#dados https://www.who.int/data/gho/data/indicators/indicator-details/GHO/total-(recorded-unrecorded)-alcohol-per-capita-(15-)-consumption
```

```python
alcohol_consumption = pd.read_csv("/content/total_alcohol_per_capita_comsuption_worldwide.csv")
alcohol_consumption
#Leitura do arquivo csv a respeito do consumo total de álcool(em litros de álcool puro)
#per capita para pessoas acima de 15 anos de vários páises do mundo.
```

```python
alcohol_consumption.info()
#Uso da função info para explorar os atributos do DataFrame e visualizar em quais
#colunas temos dados nulos e não nulos, mas também aferir quais são as colunas
#relevantes para nosso estudo.
```

```python
alcohol_consumption.drop(alcohol_consumption.iloc[:,[14,15,16,17,18,19,20,21,22,24,25,27,30]], inplace=True, axis=1)
alcohol_consumption.head(10)
#Exclusão de colunas com dados nulos.
```

```python
alcohol_consumption.info()
#checagem do novo DataFrame sem colunas com dados nulos.
```

```python
alcohol_consumption
#visualização do DataFrame após exclusão das colunas com dados nulos.
```

```python
alcohol_consumption['Period'].unique()
#verificação dos valores anuais para divisão em DataFrames com linhas de mesmo ano.
```

```python
alcohol_consumption_2019 = alcohol_consumption.loc[alcohol_consumption['Period'] == 2019]
alcohol_consumption_2015 = alcohol_consumption.loc[alcohol_consumption['Period'] == 2015]
alcohol_consumption_2010 = alcohol_consumption.loc[alcohol_consumption['Period'] == 2010]
alcohol_consumption_2005 = alcohol_consumption.loc[alcohol_consumption['Period'] == 2005]
alcohol_consumption_2000 = alcohol_consumption.loc[alcohol_consumption['Period'] == 2000]
#Separação das linhas em DataFrames com anos semelhantes.
```

```python
display(alcohol_consumption_2019,alcohol_consumption_2015,alcohol_consumption_2010,alcohol_consumption_2005,alcohol_consumption_2000)
#visualização dos DataFrames após separação.
```

#**Agora iremos selecionar os 10 maiores consumos por ano baseados no  FactValueNumeric**

```python
alcohol_consumption_2019.sort_values(by = 'FactValueNumeric', ascending = False, inplace = True)
alcohol_consumption_2015.sort_values(by = 'FactValueNumeric', ascending = False, inplace = True)
alcohol_consumption_2010.sort_values(by = 'FactValueNumeric', ascending = False, inplace = True)
alcohol_consumption_2005.sort_values(by = 'FactValueNumeric', ascending = False, inplace = True)
alcohol_consumption_2000.sort_values(by = 'FactValueNumeric', ascending = False, inplace = True)
```

```python
display(alcohol_consumption_2019,alcohol_consumption_2015,alcohol_consumption_2010,alcohol_consumption_2005,alcohol_consumption_2000)
#Visualização dos DataFrames separados por ano ordenados de forma decrescente
#com relação ao FactValueNumeric.
```

```python
alcohol_consumption_2019_Male = alcohol_consumption_2019.loc[alcohol_consumption['Dim1'] == 'Male']
alcohol_consumption_2015_Male = alcohol_consumption_2015.loc[alcohol_consumption['Dim1'] == 'Male']
alcohol_consumption_2010_Male = alcohol_consumption_2010.loc[alcohol_consumption['Dim1'] == 'Male']
alcohol_consumption_2005_Male = alcohol_consumption_2005.loc[alcohol_consumption['Dim1'] == 'Male']
alcohol_consumption_2000_Male = alcohol_consumption_2000.loc[alcohol_consumption['Dim1'] == 'Male']
alcohol_consumption_2019_Female = alcohol_consumption_2019.loc[alcohol_consumption['Dim1'] == 'Female']
alcohol_consumption_2015_Female = alcohol_consumption_2015.loc[alcohol_consumption['Dim1'] == 'Female']
alcohol_consumption_2010_Female = alcohol_consumption_2010.loc[alcohol_consumption['Dim1'] == 'Female']
alcohol_consumption_2005_Female = alcohol_consumption_2005.loc[alcohol_consumption['Dim1'] == 'Female']
alcohol_consumption_2000_Female = alcohol_consumption_2000.loc[alcohol_consumption['Dim1'] == 'Female']
alcohol_consumption_2019_Both_sexes = alcohol_consumption_2019.loc[alcohol_consumption['Dim1'] == 'Both sexes']
alcohol_consumption_2015_Both_sexes = alcohol_consumption_2015.loc[alcohol_consumption['Dim1'] == 'Both sexes']
alcohol_consumption_2010_Both_sexes = alcohol_consumption_2010.loc[alcohol_consumption['Dim1'] == 'Both sexes']
alcohol_consumption_2005_Both_sexes = alcohol_consumption_2005.loc[alcohol_consumption['Dim1'] == 'Both sexes']
alcohol_consumption_2000_Both_sexes = alcohol_consumption_2000.loc[alcohol_consumption['Dim1'] == 'Both sexes']
#Separação dos DataFrames por ano;sexo masculino, sexo feminino e ambos os sexos.
```

```python
display(alcohol_consumption_2019_Male,alcohol_consumption_2015_Male,alcohol_consumption_2019_Female,alcohol_consumption_2019_Both_sexes)
#Visualização dos DataFrames para o ano de 2019
```

```python
alcohol_consumption_2019_Male_top_10 = alcohol_consumption_2019_Male.iloc[0:10,:]
alcohol_consumption_2019_Male_top_10
#Filtragem dos 10 países em 2019 com maiores consumos totais de álcool(em litros de álcool puro)
#per capita apenas para homens acima de 15 anos.
```

```python
import matplotlib.pyplot as plt
#Importando a biblioteca matplotlib.pyplot para a plotagem de gráficos de barras
```

```python
fig, ax = plt.subplots(figsize = (17,5))
x = alcohol_consumption_2019_Male_top_10['Location']
y = alcohol_consumption_2019_Male_top_10['FactValueNumeric']
plt.title("10 maiores consumos de 2019 para homens por país baseados no FactValueNumeric")
bar_container = ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)
ax.bar_label(bar_container, fmt='{:,.0f}')
plt.show()
#Plotagem do gráfico dos 10 países em 2019 com maiores consumos totais de álcool
#(em litros de álcool puro) apenas para homens per capita acima de 15 anos.
#No eixo x temos os nomes dos países e no eixo y temos seus respectivos consumos
#totais de álcool (em litros de álcool puro) per capita apenas para homens acima
#de 15 anos.
```
